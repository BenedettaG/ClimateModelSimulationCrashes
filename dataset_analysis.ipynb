```python
infile = "/content/dati_climate.csv"
df = pd.read_csv(infile, sep=';')
from sklearn.preprocessing import StandardScaler
x = df.loc[:, features].values
x = StandardScaler().fit_transform(x) # normalizing the features
print(np.mean(x),np.std(x))


from sklearn.decomposition import PCA
pca = PCA(n_components=2, svd_solver = 'full')
principalComponents = pca.fit_transform(x)

principal_Df = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])
print(principal_Df.tail())
print('Explained variation per principal component: {}'.format(pca.explained_variance_ratio_))

pl.figure()
pl.figure(figsize=(10,10))
pl.xticks(fontsize=12)
pl.yticks(fontsize=14)
pl.xlabel('Principal Component 1',fontsize=20)
pl.ylabel('Principal Component 2',fontsize=20)
pl.title("Principal Component Analysis of Climate Model Simulation Crashes Dataset",fontsize=20)
targets = [0, 1]
colors = ['dodgerblue', 'r']
for target, color in zip(targets,colors):
    indicesToKeep = df['outcome'] == target
    pl.scatter(principal_Df.loc[indicesToKeep, 'principal component 1']
               , principal_Df.loc[indicesToKeep, 'principal component 2'], c = color, s = 50)

pl.legend(targets,prop={'size': 15})
pl.show()
```
